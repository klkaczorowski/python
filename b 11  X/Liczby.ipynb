# Uamki

```python
from fractions import Fraction
```

```python
Fraction(17, 23)   # uamek(licznik, mianownik) 
```

```python
print (Fraction(17, 23))   # licznik/manownik
```

```python
Fraction(18, 30)
```

```python
Fraction()  # domylne wartoci licznika i mianownika
```

```python
Fraction(3.14)
```

```python
Fraction(0.125)   # 1/8
```

```python
Fraction('12/18')   # mona poda acuch znakw jako argument
```

```python
print(Fraction(3,8) + Fraction(1,2))   # dodawanie uamkw
```

```python
print(Fraction(7,8) - Fraction(1,3))  # odejmowanie uamkw
```

```python
print(Fraction(7,8) * Fraction(1,3))  # mnoenie uamkw
```

```python
print(Fraction(7,8) / Fraction(1,3))  # dzielenie uamkw
```

```python
# pierwiastkowanie uamkw
import math   
print(math.sqrt(Fraction(1,4)))
```

```python
import math
print(Fraction(math.sqrt(Fraction(1,4))))
```
